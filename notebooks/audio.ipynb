# Filter

```python
from qsi.io import pre
import matplotlib.pyplot as plt

ch1 = data[:,0]
hp_filtered = pre.butter_highpass_filter(ch1, 10000, rate)

plt.figure(figsize = (24,6))
plt.plot(hp_filtered[2000:5000])
plt.title('After high-pass filtering')
plt.show()

cs.save_wav('cicada_hp.wav', hp_filtered, rate = rate, play = True)
```

```python
lp_filtered = pre.butter_lowpass_filter(ch1, 10000, rate)

plt.figure(figsize = (24,6))
plt.plot(lp_filtered[2000:5000])
plt.title('After low-pass filtering')
plt.show()

cs.save_wav('cicada_lp.wav', hp_filtered, rate = rate, play = True)
```

```python
bp_filtered = pre.filter(ch1, rate, lc=5000, hc=8000, order = 5, display = True)
```
